```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
salaries_df = pd.read_html('https://www.spotrac.com/nba/contracts/sort-value/drafted/limit-500/')[0]
```

```python
temp_df = salaries_df["Player"].str.split(" ", n = 10, expand = True)
salaries_df['First Name'] = temp_df[0]
salaries_df['Last Name'] = temp_df[1]
salaries_df.drop(columns =["Player"], inplace = True)
```

```python
salaries_df = salaries_df[['Rank', 'First Name', 'Last Name', 'Signed Age', 'Yrs', 'Value', 'AAV', 'Sign Bonus']]
    # DataFrame shows incorrectly in Jupyter Notebook
salaries_df
```

```python
engine = create_engine("sqlite:///NBA-20-21")
conn = engine.connect()
```

```python
salaries_df.to_sql('salaries', conn)
```

```python
results = engine.execute('SELECT * FROM salaries')
for r in results:
    print(r)
```
